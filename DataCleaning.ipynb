##2) Filtering the Partner observation data
Out of 7661 data - 2013 having different format:


1. Empty Values - 1948
2. Format - /observations/new?copy=\<id> = 1
4. n/a, na, none = 10
5. Weird names like Apple = 17
6. Links containing the word "Taxa" = 23
7. Format - naturalist/photos/\<id> - 1
8. \<id> = 2
9. Link starting with "1https" - 1
11. Format - observations/\<id>#activity.. = 10




```python
col = 'field:url for "partner" observation'
#Find the columns with "activity" keyword
hash_rows = df_r[col].str.contains('#', na=False)
df_r.loc[hash_rows, col] = df_r.loc[hash_rows, col].str.replace(
    r'(/observations/\d+).*', r'\1', regex=True
)
```

```python
col = 'field:url for "partner" observation'

# Change the Format - /observations/new?copy=<id> = 1 value to na
df_r[col] = df_r[col].replace('https://www.inaturalist.org/observations/new?copy=51188925', pd.NA)


#Remove the unwanted "1" character
df_r[col] = df_r[col].replace('1https://www.inaturalist.org/observations/206023212', 'https://www.inaturalist.org/observations/206023212')

#Standardize the format - /observations/<id>
df_r[col] = df_r[col].replace('97610245', 'https://www.inaturalist.org/observations/97610245')
df_r[col] = df_r[col].replace('97610244', 'https://www.inaturalist.org/observations/97610244')

#change the na or non value record to nan values
df_r[col] = df_r[col].replace('na', pd.NA)
df_r[col] = df_r[col].replace('none', pd.NA)

#Change the weird names to nan
vals = [
    "butterfly bush",
    "asciepias incarnata",
    "Acrida",
    "Haplotaxida",
    "Subterranean Termites",
    "Brachycera",
    "Holly",
    "abelha do mel",
    "Apple",
    "Flowering Plants",
    "Grasses",
    "Flowering plants",
    "Entelygyne spiders",
    "Pinks, cactuses and allies",
    "Callinectes sapidus",
    "ants",
    "Tachinid Fly"

]
df_r[col] = df_r[col].astype("string").str.strip()
df_r[col] = df_r[col].replace(vals, pd.NA)
```

```python
df_r.isna().sum()
```

```python
#Change the records containing the word "taxa" to nan
mask = df_r[col].str.contains(r'(?i)\btaxa\b', na=False)
df_r.loc[mask, col] = pd.NA
```

```python
#Drop all na values
df_r = df_r.dropna(subset=[col]).reset_index(drop=True)
df_r
```

```python
#Store the partner_IDs from the partner observation column
pattern = r'(?:observations/)?(\d+)\b'

df_r['partner_ids'] = df_r[col].apply(lambda x: re.findall(pattern, x))

df_r[['partner_ids', col]]
```

```python
#normalize the role column
role_col = 'field:id meant for "eater" or organism being eaten?'

def normalize_role(val):
    s = str(val).strip().lower()
    if 'eater' in s or 'predator' in s:
        return 'predator'
    if (('thing' in s or 'organism' in s) and 'eaten' in s) or 'prey' in s:
        return 'prey'
    return pd.NA

df_r['role'] = df_r[role_col].apply(normalize_role) if role_col else pd.NA

df_r[["id", "partner_ids", role_col, "role"]]
```

```python
# df_r['partner_ids'] = filtered_df['partner_ids'].str.split(',')
expanded_df = df_r.explode('partner_ids').reset_index(drop=True)
expanded_df.shape
```
