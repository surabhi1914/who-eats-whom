# Data Cleaning

```python
col = 'field:url for "partner" observation'
```

```python
# Strip everything after '#' (including '#')
DataDictionary[col] = DataDictionary[col].str.split('#', n=1).str[0]

#Change the Format - /observations/new?copy=<id> = 1 value to na
DataDictionary[col] = DataDictionary[col].replace('https://www.inaturalist.org/observations/new?copy=51188925', pd.NA)

#Standardize the format - /observations/<id>
DataDictionary[col] = DataDictionary[col].replace('97610245', 'https://www.inaturalist.org/observations/97610245')
DataDictionary[col] = DataDictionary[col].replace('97610244', 'https://www.inaturalist.org/observations/97610244')

##change the na or non value record to nan values
DataDictionary[col] = DataDictionary[col].replace('na', pd.NA)
DataDictionary[col] = DataDictionary[col].replace('none', pd.NA)


#Change the records containing the word "taxa" to nan; Case-insensitive word match for either 'taxa' or 'guide_taxa'
pattern = r'(?i)\b(?:guide_taxa|taxa)\b'

mask = DataDictionary[col].astype(str).str.contains(pattern, na=False)
DataDictionary.loc[mask, col] = pd.NA

#Drop the row which has word photo
DataDictionary[col] = DataDictionary[col].replace("https://www.inaturalist.org/photos/388235399", pd.NA)

#Remove the unwanted "1" character
DataDictionary[col] = DataDictionary[col].replace('1https://www.inaturalist.org/observations/206023212', 'https://www.inaturalist.org/observations/206023212')

```

```python
#Change the weird names to nan
vals = [
    "butterfly bush",
    "asciepias incarnata",
    "Acrida",
    "Haplotaxida",
    "Subterranean Termites",
    "Brachycera",
    "Holly",
    "abelha do mel",
    "Apple",
    "Flowering Plants",
    "Grasses",
    "Flowering plants",
    "Entelygyne spiders",
    "Pinks, cactuses and allies",
    "Callinectes sapidus",
    "ants",
    "Tachinid Fly"
]
# DataDictionary[col] = DataDictionary[col].astype("string").str.strip()
DataDictionary[col] = DataDictionary[col].replace(vals, pd.NA)

#Correct URL format with multiple Ids for easier splitting later
#Regex to capture iNaturalist observation URLs (case-insensitive for scheme)
URL_PATTERN = re.compile(r"https?://www\.inaturalist\.org/observations/\d+", re.IGNORECASE)
def correct_multiple_urls(entry):
    s= str(entry)
    matches = URL_PATTERN.findall(s)
    if matches:
        return ';'.join(matches)
    return s 
DataDictionary[col] = DataDictionary[col].apply(correct_multiple_urls)

DataDictionary[col]= DataDictionary[col].replace('Https://www.inaturalist.org/observations/204534071;https://www.inaturalist.org/observations/207736494', 'https://www.inaturalist.org/observations/206023212;https://www.inaturalist.org/observations/206023213')
```

```python
DataDictionary[col] = DataDictionary[col].replace('nan', pd.NA)
DataDictionary[col] = DataDictionary[col].replace('<NA>', pd.NA)

DataDictionary.isna().sum()
```

## Normalize the Role Column


```python
role_col = 'field:id meant for "eater" or organism being eaten?'
def normalize_role(val):
    s = str(val).strip().lower()
    if 'eater' in s:
        return 'predator'
    if ('thing' in s or 'organism' in s) and 'eaten' in s:
        return 'prey'
    return pd.NA

DataDictionary['role'] = DataDictionary[role_col].apply(normalize_role) if role_col else pd.NA
```
